## 흉부 암 예측 과제

Attribute Information:

1. ID number
2. Diagnosis (M = malignant, B = benign) <- Target

Ten real-valued features are computed for each cell nucleus:

1. radius (mean of distances from center to points on the perimeter)
2. texture (standard deviation of gray-scale values)
3. perimeter
4. area
5. smoothness (local variation in radius lengths)
6. compactness (perimeter^2 / area - 1.0)
7. concavity (severity of concave portions of the contour)
8. concave points (number of concave portions of the contour)
9. symmetry
10. fractal dimension ("coastline approximation" - 1)

랜덤포레스트 분류모델과 로지스틱 회귀 모델을 비교하며 과제를 수행하겠다.
1. 데이터 확인
2. 전처리 및 시각화
3. 로지스틱 회귀 모델 학습
4. 랜덤포레스트 분류 모델 학습
5. 결과 작성

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/breast_cancer.csv')
df
```

```python
df['diagnosis'] = df['diagnosis'].map({'M':1, 'B':0})
```

```python
df.info()
```

```python
df.drop('id', axis=1, inplace=True)
```

```python
df.describe()
```

```python
dfM=df[df['diagnosis'] ==1]
dfB=df[df['diagnosis'] ==0]
plt.rcParams.update({'font.size':8})
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(8,10))
axes = axes.ravel()
features_mean = list(df.columns[1:11])
binwidth= (max(df[features_mean[idx]]) - min(df[features_mean[idx]]))/50
for idx, ax in enumerate(axes):
    ax.figure
    binwidth= (max(df[features_mean[idx]]) - min(df[features_mean[idx]]))/50
    ax.hist([dfM[features_mean[idx]],dfB[features_mean[idx]]], bins=np.arange(min(df[features_mean[idx]]), max(df[features_mean[idx]]) + binwidth, binwidth), alpha=0.5, stacked=True, label=['M','B'],color=['r','g'])
    ax.set_title(features_mean[idx])
    ax.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

train_df, test_df = train_test_split(df, train_size=0.8)
features = ['radius_mean', 'perimeter_mean', 'area_mean', 'compactness_mean', 'concavity_mean']

scaler = MinMaxScaler()
train_x = scaler.fit_transform(train_df[features])
train_y = train_df['diagnosis']
test_x = scaler.fit_transform(test_df[features])
test_y = test_df['diagnosis']

lrmodel = LogisticRegression()
lrmodel.fit(train_x, train_y)

print('lrmodel train accuracy: ', lrmodel.score(train_x, train_y))
print('lrmodel test accuracy: ', lrmodel.score(test_x, test_y))
print('lrmodel coef: ', lrmodel.coef_)
```

```python
scores = []
for i in range(50, 100, 5):
    rfmodel = RandomForestClassifier(n_estimators=i, max_depth=5)
    rfmodel.fit(train_x, train_y)
    scores.append(rfmodel.score(test_x, test_y))
    
sns.lineplot(x=range(50, 100, 5), y=scores)
plt.show()
```
