```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(df['output'])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig, axs = plt.subplots(2,2, figsize=(8,8))
ax1, ax2, ax3, ax4 = axs.flatten()
ax1 = sns.pointplot(x=df['output'], y=df['cp'], ax=ax1)
ax2 = sns.pointplot(x=df['output'], y=df['trtbps'], ax=ax2)
ax3 = sns.pointplot(x=df['output'], y=df['chol'], ax=ax3)
ax4 = sns.pointplot(x=df['output'], y=df['fbs'], ax=ax4)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig, axs = plt.subplots(2,2, figsize=(8,8))
ax1, ax2, ax3, ax4 = axs.flatten()
ax1 = sns.pointplot(x=df['output'], y=df['restecg'], ax=ax1)
ax2 = sns.pointplot(x=df['output'], y=df['thalachh'], ax=ax2)
ax3 = sns.pointplot(x=df['output'], y=df['exng'], ax=ax3)
ax4 = sns.pointplot(x=df['output'], y=df['oldpeak'], ax=ax4)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

features = df[['cp', 'thalachh', 'exng', 'oldpeak']]
result = df[['output']]

train_x, test_x, train_y, test_y = train_test_split(features, result, train_size=0.8)

scaler = StandardScaler()
train_features = scaler.fit_transform(train_x)
test_features = scaler.fit_transform(test_x)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(train_features, train_y)
print(model.score(train_features, train_y))
print(model.score(test_features, test_y))
print(model.coef_)
```

```python
df[['cp', 'thalachh', 'exng', 'oldpeak']].describe()
```

```python
import numpy as np
heejin = np.array([2.1, 170, 0.3, 1.3])
heejoo = np.array([0, 135, 0, 0.0])
soon = np.array([2, 166, 1, 1.6])
mung = np.array([1.7, 130, 1.2, 1.8])
sample = np.array([heejin, heejoo, soon, mung])
sample = scaler.transform(sample)
print(model.predict(sample))
print(model.predict_proba(sample))
```
