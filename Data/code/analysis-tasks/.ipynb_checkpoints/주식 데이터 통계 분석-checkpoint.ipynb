```python
brand_df = df.groupby('Brand_Name')['High'].mean()
brand_df = pd.DataFrame(brand_df)
brand_df
```

```python
def level(x):
    if x > brand_df['High'].describe()['75%']:
        return 1
    elif x <=  brand_df['High'].describe()['75%'] and x > brand_df['High'].describe()['50%']:
        return 2
    elif x <=  brand_df['High'].describe()['50%'] and x >  brand_df['High'].describe()['25%']:
        return 3
    else:
        return 4
    
brand_df['level'] = brand_df['High'].apply(lambda x: level(x))
```

```python
brand_df
```

```python
ddd = df.merge(brand_df, on='Brand_Name', how='left')
```

```python
df = pd.read_csv('../data/World-Stock-Prices-Dataset.csv')
df['Country'].value_counts()
df['Country'].count()
```

```python
ratio_df = pd.DataFrame(df['Country'].value_counts()/df['Country'].count())
ratio_df.reset_index(inplace=True)
ratio_df
```

```python
plt.figure(figsize = (10,10))
plt.pie(data=ratio_df, x='Country', labels='index', autopct='%.2f%%')
plt.show()
```
